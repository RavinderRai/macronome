```python
from PIL import Image

# Load the image with transparency
input_path = "adaptive-icon-removebg-preview.png"
output_path = "adaptive-icon.png"

# Open the image
img = Image.open(input_path)

# Check if it has an alpha channel (transparency)
if img.mode != 'RGBA':
    img = img.convert('RGBA')

# Resize to 1024x1024 using high-quality LANCZOS resampling
# This preserves transparency and maintains quality
resized_img = img.resize((1024, 1024), Image.Resampling.LANCZOS)

# Save as PNG to preserve transparency
resized_img.save(output_path, 'PNG', optimize=False)

print(f"✓ Resized {input_path} from {img.size} to {resized_img.size}")
print(f"✓ Saved as {output_path}")
print(f"✓ Transparency preserved: {resized_img.mode == 'RGBA'}")
```

```python
# Also create icon.png (for iOS/general use) from the same source
icon_output_path = "icon.png"

# Load and resize again (or reuse the resized_img from previous cell)
img_for_icon = Image.open(input_path)
if img_for_icon.mode != 'RGBA':
    img_for_icon = img_for_icon.convert('RGBA')
icon_resized = img_for_icon.resize((1024, 1024), Image.Resampling.LANCZOS)
icon_resized.save(icon_output_path, 'PNG', optimize=False)

print(f"✓ Also created {icon_output_path} at 1024x1024")
```
